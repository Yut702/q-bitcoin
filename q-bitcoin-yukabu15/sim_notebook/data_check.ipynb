```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
# SPSA optimizer
from qiskit_algorithms.optimizers import SPSA
import math
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

# 数値安定化用の上限値
MAX_WEIGHT = 2.0
MAX_SCALE = 2.0
MAX_BIAS = 2.0

# Aer backend for shot-based simulation
backend = AerSimulator()

# --- データ読み込み ---
df = pd.read_excel('../data/Dataset_Simulated_Price_swaption.xlsx')  # ここだけ変更していただければ動くと思います

print('data shape =', df.shape)
display(df.head())

# 数値列のみを対象にする（非数値列は除外）
df_numeric = df.select_dtypes(include=[np.number]).copy()
if df_numeric.shape[1] < df.shape[1]:
    dropped_cols = [col for col in df.columns if col not in df_numeric.columns]
    print('Non-numeric columns dropped for modeling:', dropped_cols)
if df_numeric.isnull().values.any():
    df_numeric = df_numeric.fillna(method='ffill').fillna(method='bfill')
    if df_numeric.isnull().values.any():
        df_numeric = df_numeric.fillna(0.0)
    print('Missing values were filled (ffill/bfill).')

target_columns = df_numeric.columns.tolist()
print('Modeling columns:', target_columns)

data_values = df_numeric.values

# ウィンドウ化（すべての列をまとめて扱う）
window = 7
X_seq = []
y_seq = []
for i in range(len(data_values) - window):
    X_seq.append(data_values[i:i+window])
    y_seq.append(data_values[i+window])
X_seq = np.array(X_seq)  # (samples, window, features)
y = np.array(y_seq)      # (samples, features)
X = X_seq.reshape(X_seq.shape[0], -1)  # (samples, window*features)

print('X shape (samples, window*features) =', X.shape)
print('y shape (samples, targets) =', y.shape)

# スケーリング（特徴量とターゲットそれぞれ個別に標準化）
x_scaler = StandardScaler()
X_scaled = x_scaler.fit_transform(X)
y_scaler = StandardScaler()
y_scaled = y_scaler.fit_transform(y)

# PCA -> n_qubits 次元に削減して n_qubits qubits 用にする
n_qubits = 4
pca = PCA(n_components=n_qubits)
Xred = pca.fit_transform(X_scaled)
print('PCA explained variance ratio:', pca.explained_variance_ratio_)

# 角度にスケール（0..2pi）
def scale_to_angle(arr, low=0.0, high=2*math.pi):
    amin = arr.min(axis=0)
    amax = arr.max(axis=0)
    rng = np.where(amax - amin == 0, 1e-6, amax - amin)
    return low + (arr - amin) / rng * (high - low)

Xangle = scale_to_angle(Xred)
print('Xangle range per dim:', Xangle.min(axis=0), Xangle.max(axis=0))

# train/test split
X_train, X_test, y_train, y_test = train_test_split(Xangle, y_scaled, test_size=0.3, random_state=42)
print('train/test sizes:', X_train.shape[0], X_test.shape[0])

# 量子回路定義（n_qubits qubits）
n_layers = 2
n_targets = y_train.shape[1]
print('Number of qubits =', n_qubits, ', Number of targets =', n_targets)

def feature_encoding_circuit(features):
    qc = QuantumCircuit(n_qubits)
    for i, val in enumerate(features):
        qc.ry(val, i)
    return qc

def ansatz_circuit(params, n_layers=2):
    # params length should be 2 * n_qubits * n_layers
    qc = QuantumCircuit(n_qubits)
    idx = 0
    for L in range(n_layers):
        for i in range(n_qubits):
            qc.rx(params[idx], i); idx += 1
            qc.ry(params[idx], i); idx += 1
        # entangling chain
        for i in range(n_qubits - 1):
            qc.cx(i, i+1)
    return qc

def full_circuit(features, params):
    qc = feature_encoding_circuit(features)
    qc = qc.compose(ansatz_circuit(params, n_layers=n_layers))
    return qc

def unpack_params(params):
    n_circ = 2 * n_qubits * n_layers
    circ_params = params[:n_circ]
    w_start = n_circ
    w_end = w_start + n_targets * n_qubits
    scale_start = w_end
    scale_end = scale_start + n_targets
    bias_start = scale_end
    bias_end = bias_start + n_targets
    w_raw = params[w_start:w_end].reshape(n_targets, n_qubits)
    scale_raw = params[scale_start:scale_end]
    bias_raw = params[bias_start:bias_end]
    w = np.tanh(w_raw) * MAX_WEIGHT
    scale = np.tanh(scale_raw) * MAX_SCALE
    bias = np.tanh(bias_raw) * MAX_BIAS
    return circ_params, w, scale, bias

# 各 qubit の Z 期待値をカウントから推定（shots ベース）
def circuit_expectations_shots_batch(features_batch, params_circ, shots=1000, progress_desc=None):
    """複数入力をまとめてバックエンドに投げ、一括で期待値を推定する"""
    features_array = np.atleast_2d(features_batch)
    circuits = []
    for feats in features_array:
        qc = full_circuit(feats, params_circ)
        qc_measure = qc.copy()
        qc_measure.measure_all()
        circuits.append(qc_measure)
    t_qcs = transpile(circuits, backend)
    job = backend.run(t_qcs, shots=shots)
    result = job.result()
    exps_batch = []
    for idx in range(len(circuits)):
        counts = result.get_counts(idx)
        total = sum(counts.values()) if counts else 0
        exps = np.zeros(n_qubits)
        if total == 0:
            exps_batch.append(exps)
            continue
        for bitstr, cnt in counts.items():
            p = cnt / total
            for q_idx in range(n_qubits):
                b = int(bitstr[-1 - q_idx])
                val = 1 if b == 0 else -1
                exps[q_idx] += val * p
        exps_batch.append(np.real(exps))
    return np.array(exps_batch)

def circuit_expectations_shots(features, params_circ, shots=1000):
    return circuit_expectations_shots_batch(np.asarray(features), params_circ, shots=shots)[0]

# predict: params 配列を分割して circ_params, weight matrix, scale vector, bias vector を取り出して計算
# params format: [circ_params (2*n_qubits*n_layers), w (n_targets*n_qubits), scale (n_targets), bias (n_targets)]
def predict_batch_from_params(features_batch, params, shots=1000, progress_desc=None):
    features_array = np.atleast_2d(features_batch)
    circ_params, w, scale, bias = unpack_params(params)
    exps_batch = circuit_expectations_shots_batch(features_array, circ_params, shots=shots, progress_desc=progress_desc)
    linear = exps_batch.dot(w.T)  # (samples, n_targets)
    return linear * scale + bias

def predict_from_params(features, params, shots=1000):
    return predict_batch_from_params(np.asarray(features), params, shots=shots)[0]

# loss 履歴を記録するリスト
loss_history = []

def loss_fn(params):
    preds = predict_batch_from_params(X_train, params, shots=1000, progress_desc='SPSA train circuits')
    # NaN/inf ガード
    if not np.all(np.isfinite(preds)):
        loss_history.append(1e6)
        return 1e6
    loss = np.mean((preds - y_train)**2)
    loss_history.append(loss)
    return loss

# パラメータ初期化: circ params + weight matrix + scale + bias
n_circ = 2 * n_qubits * n_layers
param_size = n_circ + n_targets * (n_qubits + 2)
init_params = np.random.normal(0, 0.1, size=param_size)
w_start = n_circ
w_end = w_start + n_targets * n_qubits
scale_start = w_end
scale_end = scale_start + n_targets
bias_start = scale_end
bias_end = bias_start + n_targets
init_params[scale_start:scale_end] = 0.1
init_params[bias_start:bias_end] = 0.0

print('param vector size =', param_size)

maxiter = 100  # 既存の設定に合わせる
progress_bar = tqdm(total=maxiter, desc="SPSA iterations", unit="iter")

def spsa_callback(evals, params, loss, stepsize, accepted):
    # tqdmは1-based更新でOK。過剰更新を防ぎたい場合はmin(maxiter, evals)との差分だけ進める
    progress_bar.update(1)
    progress_bar.set_postfix({"loss": f"{loss:.4f}", "accepted": bool(accepted)})

# SPSA による最適化 (反復回数は必要に応じて増やすが、shots=1000 は遅いので注意)
opt = SPSA(maxiter=maxiter, callback=spsa_callback)
res = opt.minimize(loss_fn, x0=init_params)
params_opt = np.array(res.x)
print('Optimized parameters obtained.')

# テスト評価（スケールを元に戻して評価）
# ここも shots=1000 を使って評価
preds_test = predict_batch_from_params(X_test, params_opt, shots=1000, progress_desc='Evaluating test circuits')
print('preds_test (scaled) range per column:')
display(pd.DataFrame({'min': preds_test.min(axis=0), 'max': preds_test.max(axis=0)}, index=target_columns))

y_test_orig = y_scaler.inverse_transform(y_test)
preds_test_orig = y_scaler.inverse_transform(preds_test)
print('preds_test (orig scale) range per column:')
display(pd.DataFrame({'min': preds_test_orig.min(axis=0), 'max': preds_test_orig.max(axis=0)}, index=target_columns))

mse_test_scaled = np.mean((preds_test - y_test)**2, axis=0)
mse_test_orig = np.mean((preds_test_orig - y_test_orig)**2, axis=0)
print('test MSE (scaled) per column:')
display(pd.Series(mse_test_scaled, index=target_columns))
print('test MSE (scaled) mean =', float(mse_test_scaled.mean()))
print('test MSE (original scale) per column:')
display(pd.Series(mse_test_orig, index=target_columns))
print('test MSE (original scale) mean =', float(mse_test_orig.mean()))

# loss 履歴の可視化
plt.figure(figsize=(6, 3))
plt.plot(loss_history, label='loss (per eval)')
plt.yscale('log')
plt.xlabel('evaluation index')
plt.ylabel('MSE (scaled target)')
plt.legend()
plt.title('Loss history during optimization')
plt.show()

# 真値 vs 予測プロット（代表として最初の列を表示）
target_idx_to_plot = 10
target_name_to_plot = target_columns[target_idx_to_plot] if target_columns else 'target0'
plt.figure(figsize=(6, 3))
plt.plot(y_test_orig[:, target_idx_to_plot], label=f'true (orig) {target_name_to_plot}')
plt.plot(preds_test_orig[:, target_idx_to_plot], label=f'pred (orig) {target_name_to_plot}')
plt.legend()
plt.title(f'True vs Pred (orig scale) [{target_name_to_plot}]')
plt.show()
```
